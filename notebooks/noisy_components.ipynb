```python
import pickle
import sys
import os
import mne
# Add the workspace root to the Python path
workspace_root = os.path.abspath("..")
if workspace_root not in sys.path:
    sys.path.append(workspace_root)
from functions import  get_noisyICs
import matplotlib.pyplot as plt
ids = [601,602,603,604,605,606]
for id in ids:

    path_labels = os.path.join(workspace_root,'Data', f'S{id}_ica_infomax_LpF_labels.pkl')
    with open(path_labels, "rb") as f:
        ic_labels = pickle.load(f)
    ica_path = os.path.join(workspace_root, 'Data', f'S{id}_ica_infomax_LpF.fif')
    ica = mne.preprocessing.read_ica(ica_path)
    noisy_components = get_noisyICs(ic_labels, threshold= 0.7, noise_type='all')
    for i, label in enumerate(ic_labels['labels']):
            if i in noisy_components:
                prob = ic_labels['y_pred_proba'][i]
                print(f'component {i} is {label} with probability {prob:.2f}')
    plt.figure(figsize=(10, 5))
    ica.plot_components( picks = noisy_components, title = f'noisy components for subject {id}', show =False )
    plt.show()
```
